```python
import datetime as dt
import os
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns

today = dt.datetime.today().strftime("%Y-%m-%d")
data_folder = "/Users/annasintsova/git_repos/HUTI-RNAseq/results/metabolic_models/"
files = [os.path.join(data_folder, f) for f in os.listdir(data_folder)]
```

```python
r_names = pd.read_csv(files[0], sep = '\t', index_col = 0, names = ["Reaction_Name", "conf"]).Reaction_Name
```

```python
def read_file(file, fmt = "tab", today = today):
    name = os.path.basename(file).split(".")[0].strip(today+"_")
    if fmt == "tab":
        df = pd.read_csv(file, sep="\t", index_col = 0, names = ["Reaction_name", name])
        ph = df.Reaction_name
        #print(ph)
        df.drop("Reaction_name", axis = 1,inplace=True)
    return df
```

```python
file = files[0]
df = pd.concat([read_file(file) for file in files], axis=1)
df2 = pd.concat([df, r_names], axis =1)
df2
```

```python
df = df.dropna()
```

```python
df = df[df.sum(axis=1)!=0]
sns.clustermap(df)
```

```python
uti_cols = [c for c in df.columns if "UTI" in c]
ur_cols = [c for c in df.columns if "UR" in c]
uti_df = df[uti_cols]
ur_df = df[ur_cols]
```

```python
uti_hi = uti_df[uti_df.sum(axis=1) == len(uti_cols)*2]
urine_low = ur_df[ur_df.sum(axis=1) < len(uti_cols)*2]

uti_sp = set(uti_hi.index).intersection(set(urine_low.index))
uti_sp
```
