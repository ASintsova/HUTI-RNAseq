# Merging the count data

```python
import pandas as pd
```

```python
strain_nums = [1, 3, 6, 7, 14, 17, 43, 54, 56, 57, 66, 68, 86 ]
media = ['UR', 'UTI']
filenames = []
for strain in strain_nums:
    for m in media:
        s = m + str(strain)
        filenames.append(s)
filenames.remove("UR1")
```

```python
comb = pd.read_table("UR1", sep = "\t", header = None)
comb.columns = ["Gene_id", 'UR1']
comb = comb.set_index("Gene_id")
```

```python
for name in filenames:
    x = pd.read_table(name, sep = "\t", header = None)
    x.columns = ["Gene_id", name]
    x = x.set_index("Gene_id")
    comb = pd.merge(comb, x, how='outer', left_index=True, right_index=True)
print (comb.head())
```
