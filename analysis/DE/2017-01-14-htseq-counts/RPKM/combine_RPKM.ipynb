```python
import pandas as pd
strain_nums = [1, 3, 6, 7, 14, 17, 43, 54, 56, 57, 66, 68, 86 ]
media = ['UR', 'UTI']
filenames = []
for strain in strain_nums:
    for m in media:
        s = m + str(strain)+ ".RPKM"
        filenames.append(s)
filenames.remove("UR1.RPKM")
```

```python
comb = pd.read_table("UR1.RPKM", sep = "\t", header = 1)
comb.columns = ["Gene_id", 'UR1']
comb = comb.set_index("Gene_id")
```

```python
comb.head()
```

```python
for name in filenames:
    x = pd.read_table(name, sep = "\t", header = 1)
    x.columns = ["Gene_id", name[:-5]]
    x = x.set_index("Gene_id")
    comb = pd.merge(comb, x, how='outer', left_index=True, right_index=True)
print (comb.head())
```

```python
comb.to_csv("combined_RPKM.csv")
```
